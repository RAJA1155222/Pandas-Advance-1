Q2. Re-index DataFrame starting from 1 and increment by 2

```python
def reindex_df(df):
    df = df.copy()
    df.index = range(1, 1 + 2*len(df), 2)
    return df
```

Q3. Sum of first three values of column 'Values'

```python
def sum_first_three(df):
    total = 0
    for i, val in enumerate(df["Values"]):
        if i < 3:
            total += val
    print(total)
```

Q7. Extract username from email

```python
def extract_username(df):
    df = df.copy()
    df["Username"] = df["Email"].str.split("@").str[0]
    return df
```

Q8. Select rows where A > 5 and B < 10

```python
def filter_rows(df):
    return df[(df["A"] > 5) & (df["B"] < 10)]
```

Q9. Mean, median and standard deviation

```python
def stats_values(df):
    mean = df["Values"].mean()
    median = df["Values"].median()
    std = df["Values"].std()
    return mean, median, std
```

Q10. 7-day moving average of Sales

```python
def add_moving_average(df):
    df = df.copy()
    df["Date"] = pd.to_datetime(df["Date"])
    df = df.sort_values("Date")
    df["MovingAverage"] = df["Sales"].rolling(window=7).mean()
    return df
```

Q11. Create Weekday column from Date

```python
def add_weekday(df):
    df = df.copy()
    df["Date"] = pd.to_datetime(df["Date"])
    df["Weekday"] = df["Date"].dt.day_name()
    return df
```

Q12. Select rows between two dates

```python
def filter_date_range(df):
    df = df.copy()
    df["Date"] = pd.to_datetime(df["Date"])

    return df[
        (df["Date"] >= "2023-01-01") &
        (df["Date"] <= "2023-01-31")
    ]
```
